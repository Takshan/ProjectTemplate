# Preperation of file for custom scoring function

Need to mannually add the columns name from any of the log file.

```python
cols =['name','gauss(o=0,_w=0.5,_c=8)', 'gauss(o=3,_w=2,_c=8)', 'repulsion(o=0,_c=8)', 'hydrophobic(g=0.5,_b=1.5,_c=8)','non_dir_h_bond(g=-0.7,_b=0,_c=8)', 'num_tors_div', 'atom_type_gaussian(t1=Chlorine,t2=Sulfur,o=0,_w=3,_c=8)', "Affinity"]
len(cols)
```

Parse the log file to prepare for the requiremnet for further step.

```python
temp_df = pd.DataFrame(columns=cols)
for count,i in enumerate(score_files):
    _affinity = []
    term_values = []
    temp_dir, temp_name, temp_format = give_id(i)
    with open(i, "r") as rd:
        for line in rd:
            if line.startswith("## Name"):
                if len(cols) <1:
                    cols.extend(line[7:].split())
            elif line.startswith("Affinity:"):
                _affinity.append(line.split()[1])
            elif line.startswith("##  "):
                term_values.append(temp_name)
                term_values.extend(line.split()[1:])
                term_values.extend(_affinity)

    temp_df.loc[count] = [x for x in term_values]

temp_df
```

### For active and inactive adding new column "STATE" as 0 or 1 for inactive and active

sample making all default as active 

```python
content = ["default", "inactive"] # marker to select active and inactive
value = [1, 0] # value for the marker 
content_list = [temp_df.name.str.lower().str.contains(con) for con in content]
temp_df['State'] = np.select(content_list, value)
temp_df
```
